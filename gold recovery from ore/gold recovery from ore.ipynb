### Проведка адекватности данных

Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака rougher.output.recovery. Найдем MAE между расчётами и значением признака.

Для упрощения формулы, объявим доли константами.

```python
C = df_train['rougher.output.concentrate_au'] # Доля золота после флотации
F = df_train['rougher.input.feed_au'] # Доля золота до флотации
T = df_train['rougher.output.tail_au'] # Доля золота в хвостах после флотации
```

```python
true_recovery = (C * (F - T)) / (F * (C - T)) * 100
```

```python
mae = mean_absolute_error(true_recovery, df_train['rougher.output.recovery'])
mae
```

**Вывод:** эффективность обогащения рассчитана правильно, ошибка достигает погрешности вычислений.

### Изучение суммарной концентрации веществ на разных стадиях

На обучающей выборке посмотрим, как менялась суммарная концентрация всех веществ на разных стадиях. 

Вначале рассчитаем сумарную концентрацию всех веществ.

```python
df_train['rougher.input.feed_sum'] = df_train[[
    'rougher.input.feed_au',
    'rougher.input.feed_ag',
    'rougher.input.feed_pb',
    'rougher.input.feed_sol'
]].sum(axis=1)

df_train['rougher.output.concentrate_sum'] = df_train[[
    'rougher.output.concentrate_au',
    'rougher.output.concentrate_ag',
    'rougher.output.concentrate_pb',
    'rougher.output.concentrate_sol'
]].sum(axis=1)

df_train['primary_cleaner.output.concentrate_sum'] = df_train[[
    'primary_cleaner.output.concentrate_au',
    'primary_cleaner.output.concentrate_ag',
    'primary_cleaner.output.concentrate_pb',
    'primary_cleaner.output.concentrate_sol'
]].sum(axis=1)
    
df_train['final.output.concentrate_sum'] = df_train[[
    'final.output.concentrate_au',
    'final.output.concentrate_ag',
    'final.output.concentrate_pb',
    'final.output.concentrate_sol'
]].sum(axis=1)
```

```python
ax = (
    df_train[['rougher.input.feed_sum', 'rougher.output.concentrate_sum',
              'primary_cleaner.output.concentrate_sum','final.output.concentrate_sum']]
    .plot
    .hist(bins=30, alpha=0.5)
);

ax.set(
    xlabel='Концентрация веществ, %',
    ylabel='Частота',
    title='Распределение концентрации веществ на разных этапах'
)
plt.legend(['Сырье', 'Флотация', 'Перв. очистка', 'Втор. очистка'])
plt.show()
```

**Вывод:** суммарная концентрация веществ меняется, причем максимальна она в финальном концентрате, а в сырье минимальная.

На графике заметны выбросы - все значения меньше 20% явно стоит удалить.

## Модель

Для начала объявим функцию для вычисления требуемой метрики - итоговой sMAPE `res_smape`.

### Функция для sMAPE

```python
def smape(true, pred):
    return 100 / len(true) * np.sum(np.abs(true - pred) /  (np.abs(true) + np.abs(pred) / 2))

def res_smape(smape_rougher, smape_final):
    return 0.25 * smape_rougher + 0.75 * smape_final
```
