## Исследование задачи

Исследуем баланс классов, обучим модели без учета дисбаланса, сделаем выводы.

Воспользуемся логистической регрессией, решающим деревом, случайным лесом и, ради эксперимента, методом опорных векторов с радиальной базисной функцией.

### Обучение без учета дисбаланса классов

Создадим функцию, которая выводит на экран сразу все метрики классификации.

```python
def classification_metrics(target, predictions, include_accuracy=True):
    '''Выводит на экран различные метрики классификации'''
    if include_accuracy:
        print('Доля правильных ответов\n', accuracy_score(target, predictions))
    print('Матрица ошибок\n', confusion_matrix(target, predictions))
    print('Полнота\n', recall_score(target, predictions))
    print('Точность\n', precision_score(target, predictions))
    print('F1-мера\n', f1_score(target, predictions))
```

#### Логистическая регрессия (LR)

```python
%%time

lr_model = LogisticRegression(random_state=345)
lr_model.fit(features_train, target_train)
pred_lr = lr_model.predict(features_valid)
```

```python
classification_metrics(target_valid, pred_lr)
```

```python
probabilities_valid = lr_model.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]
print('AUC-ROC:', roc_auc_score(target_valid, probabilities_one_valid))
```

**Вывод:** при довольно высокой доле правильных ответов, LR модель выдает значительное количество ложноотрицательных ответов. Без учета дисбаланса классов F1-мера всего 0.503. При этом, модель довольно адекватная, AUC-ROC более 0.831.

#### Решающее дерево (DT)

Попробуем сразу максимизировать f1-меру перебором гиперпараметра `max_depth`.

```python
%%time

test_dt = pd.DataFrame(columns=['max_depth', 'f1_score', 'auc_roc'])
f1_dt = 0
best_depth = 0
f1_valid = []
f1_train = []

for depth in tqdm(range(1, 51)):
    dt_model = DecisionTreeClassifier(max_depth=depth, random_state=345)
    dt_model.fit(features_train, target_train)
    pred_dt = dt_model.predict(features_valid)
    pred_train_dt = dt_model.predict(features_train)
    
    probabilities_valid = dt_model.predict_proba(features_valid)
    probabilities_one_valid = probabilities_valid[:, 1]
    
    if f1_score(target_valid, pred_dt) > f1_dt:
        f1_dt = f1_score(target_valid, pred_dt)
        best_depth = depth
    test_dt = test_dt.append(
        pd.DataFrame(
            [[depth, f1_score(target_valid, pred_dt), roc_auc_score(target_valid, probabilities_one_valid)]],
            columns=['max_depth', 'f1_score', 'auc_roc']),
        ignore_index=True)
    f1_valid.append(f1_score(target_valid, pred_dt))
    f1_train.append(f1_score(target_train, pred_train_dt))
```

```python
test_dt.sort_values(by='f1_score', ascending=False).head()
```

```python
dt_model_best = DecisionTreeClassifier(max_depth=best_depth, random_state=345)
dt_model_best.fit(features_train, target_train)
pred_dt_best = dt_model_best.predict(features_valid)
print('Лучшая глубина:', best_depth)
classification_metrics(target_valid, pred_dt_best)
```

### Upsampling

Вначале объявим функцию для увеличения числа объектов целевого признака в выборке и произведем upsampling в 3 раза (чтобы довести соотношение примерно до 1:1).

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=345)
    return features_upsampled, target_upsampled
```

```python
features_upsampled, target_upsampled = upsample(features_train, target_train, 3)
```

#### Логистическая регрессия

```python
lr_model = LogisticRegression(random_state=345, solver='liblinear')
lr_model.fit(features_upsampled, target_upsampled)
pred_lr = lr_model.predict(features_valid)
classification_metrics(target_valid, pred_lr)
```

F1-мера увеличилась с 0.503 до 0.574, что выше, чем при взвешивании.

#### Решающее дерево

### Downsampling

Вначале объявим функцию для уменьшения числа объектов целевого признака в выборке и произведем downsampling в 4 раза (чтобы довести соотношение примерно до 1:1).

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    features_downsampled = pd.concat(
    [features_zeros.sample(frac=fraction, random_state=345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=345)] + [target_ones])
    features_downsampled, target_downsampled = shuffle(
    features_downsampled, target_downsampled, random_state=345)
    return features_downsampled, target_downsampled
```

Это было высокое значение F1, а также модель имеет наибольшее значение `AUC-ROC: 0.856` поэтому выбираем для тестирования RF модель с параметрами `n_estimators = 45`и `max_depth = 7`и используем для борьбы с дисбалансом изменение порога с порогом 0.3.

## Тестирование модели
